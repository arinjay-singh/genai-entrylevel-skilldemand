## Create Panel Data

```python
import pandas as pd
from itertools import product

# Combine yearly dataframes
panel = pd.concat([df_2021, df_2022, df_2023, df_2024], ignore_index=True)

# Get all unique skills and occupations across all years
all_skills = panel['skill'].unique()
all_occs = panel['occupation_group'].unique()
all_years = [2021, 2022, 2023, 2024]

# Create balanced panel with all skill-occupation-year combinations
full_index = pd.DataFrame(
    list(product(all_occs, all_skills, all_years)),
    columns=['occupation_group', 'skill', 'year']
)

# Merge and fill missing with 0
panel = full_index.merge(panel, on=['occupation_group', 'skill', 'year'], how='left')
panel['pct'] = panel['pct'].fillna(0)
panel['count'] = panel['count'].fillna(0).astype(int)

# Add treatment and post indicators
panel['treated'] = panel['skill'].isin(high_ai_sub).astype(int)
panel['post'] = (panel['year'] >= 2023).astype(int)

print(f"Panel shape: {panel.shape}")  # should be 21,204 x columns
```

## Check Variance Inflation Factor for Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

# Build the design matrix for your main regression
# (same variables you'll use in your DiD)
vif_data = pd.get_dummies(panel, columns=['occupation_group', 'year'], drop_first=True)

# Keep only the regression variables
reg_cols = (['treated', 'treated_post'] +
            [c for c in vif_data.columns if c.startswith('occupation_group_')] +
            [c for c in vif_data.columns if c.startswith('year_')])

X = vif_data[reg_cols].astype(float)
X = X.dropna()

# Calculate VIF for each variable
vif_df = pd.DataFrame({
    'Variable': X.columns,
    'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
}).sort_values('VIF', ascending=False)

print(vif_df)
```

# Regression Results

```python
import statsmodels.formula.api as smf

# Main specification - clustered SE
model_clustered = smf.ols(
    'pct ~ treated + treated_post + C(occupation_group) + C(year)',
    data=panel
).fit(cov_type='cluster', cov_kwds={'groups': panel['occupation_group']})

# Robustness - HC3 heteroscedasticity-robust SE
model_hc3 = smf.ols(
    'pct ~ treated + treated_post + C(occupation_group) + C(year)',
    data=panel
).fit(cov_type='HC3')

# Compare the two
print("Clustered SE:")
print(f"treated_post coef: {model_clustered.params['treated_post']:.4f}")
print(f"treated_post p-value: {model_clustered.pvalues['treated_post']:.4f}")

print("\nHC3 SE:")
print(f"treated_post coef: {model_hc3.params['treated_post']:.4f}")
print(f"treated_post p-value: {model_hc3.pvalues['treated_post']:.4f}")

print(f"\nR-squared: {model_clustered.rsquared:.4f}")
```

## Redefining Post as 2024 Only

## Per Occupation Group Regressions

Post = 2024

```python
results = []

for occ in panel['occupation_group'].unique():
    occ_data = panel[panel['occupation_group'] == occ]

    try:
        model_occ = smf.ols(
            'pct ~ treated + treated_post_2024 + C(year)',
            data=occ_data
        ).fit(cov_type='HC3')

        results.append({
            'occupation': occ,
            'beta3': model_occ.params['treated_post_2024'],
            'se': model_occ.bse['treated_post_2024'],
            'pvalue': model_occ.pvalues['treated_post_2024'],
            'nobs': int(model_occ.nobs),
            'rsquared': model_occ.rsquared
        })

    except Exception as e:
        print(f"Failed for {occ}: {e}")

# Display as sorted table
results_df = pd.DataFrame(results).sort_values('beta3')
results_df['beta3'] = results_df['beta3'].round(3)
results_df['se'] = results_df['se'].round(3)
results_df['pvalue'] = results_df['pvalue'].round(3)
results_df['rsquared'] = results_df['rsquared'].round(3)

print(results_df.to_string(index=False))
```

Post as Original Definition

```python
results_original = []

for occ in panel['occupation_group'].unique():
    occ_data = panel[panel['occupation_group'] == occ]

    try:
        model_occ = smf.ols(
            'pct ~ treated + treated_post + C(year)',
            data=occ_data
        ).fit(cov_type='HC3')

        results_original.append({
            'occupation': occ,
            'beta3': model_occ.params['treated_post'],
            'se': model_occ.bse['treated_post'],
            'pvalue': model_occ.pvalues['treated_post'],
            'nobs': int(model_occ.nobs),
            'rsquared': model_occ.rsquared
        })

    except Exception as e:
        print(f"Failed for {occ}: {e}")

results_original_df = pd.DataFrame(results_original).sort_values('beta3')
results_original_df['beta3'] = results_original_df['beta3'].round(3)
results_original_df['se'] = results_original_df['se'].round(3)
results_original_df['pvalue'] = results_original_df['pvalue'].round(3)
results_original_df['rsquared'] = results_original_df['rsquared'].round(3)

print(results_original_df.to_string(index=False))
```

```python
# Create year interaction terms
panel['treated_2022'] = panel['treated'] * (panel['year'] == 2022).astype(int)
panel['treated_2023'] = panel['treated'] * (panel['year'] == 2023).astype(int)
panel['treated_2024'] = panel['treated'] * (panel['year'] == 2024).astype(int)

# Event study regression
model_event = smf.ols(
    'pct ~ treated + treated_2022 + treated_2023 + treated_2024 + C(occupation_group) + C(year)',
    data=panel
).fit(cov_type='cluster', cov_kwds={'groups': panel['occupation_group']})

print(f"treated_2022: {model_event.params['treated_2022']:.4f}, p = {model_event.pvalues['treated_2022']:.4f}")
print(f"treated_2023: {model_event.params['treated_2023']:.4f}, p = {model_event.pvalues['treated_2023']:.4f}")
print(f"treated_2024: {model_event.params['treated_2024']:.4f}, p = {model_event.pvalues['treated_2024']:.4f}")
```

```python
def add_stars(pvalue):
    if pvalue < 0.01:
        return '***'
    elif pvalue < 0.05:
        return '**'
    elif pvalue < 0.10:
        return '*'
    else:
        return ''

print(f"{'':25} {'Base':>12} {'Occ FE':>12} {'Year FE':>12} {'Two-Way FE':>12}")
print("-" * 73)

for var, label in [('treated', 'β₁ (Treated)'),
                    ('treated_2022', 'Treated × 2022'),
                    ('treated_2023', 'Treated × 2023'),
                    ('treated_2024', 'Treated × 2024')]:
    # Coefficient row with stars
    row = f"{label:25}"
    for m in event_models:
        stars = add_stars(m.pvalues[var])
        row += f"{str(round(m.params[var], 3)) + stars:>12}"
    print(row)
    # SE row
    row = f"{'':25}"
    for m in event_models:
        row += f"{'(' + f'{m.bse[var]:.3f}' + ')':>12}"
    print(row)

print("-" * 73)
print(f"{'R²':25}" + "".join(f"{m.rsquared:>12.3f}" for m in event_models))
print(f"{'Occupation FE':25} {'No':>12} {'Yes':>12} {'No':>12} {'Yes':>12}")
print(f"{'Year FE':25} {'No':>12} {'No':>12} {'Yes':>12} {'Yes':>12}")
print(f"{'N':25} {'21,204':>12} {'21,204':>12} {'21,204':>12} {'21,204':>12}")
print("\n* p < 0.10, ** p < 0.05, *** p < 0.01")
print("Standard errors clustered at occupation level in parentheses")
```
